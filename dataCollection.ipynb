```python
import pandas as pd
```

**Read in the user.json file, remove uneccesary columns and convert it to a csv file to reduce overhead**

```python
user = pd.read_json('dataset/user.json', lines=True, chunksize=1000)

df_user = pd.DataFrame()

for chunk in user:
    df_user = pd.concat([df_user,chunk])
```

```python
df_user.head()
```

```python
df_user = df_user[['user_id','average_stars','review_count','useful']]
df_user.head()
```

```python
df_user.to_csv('dataset/user.csv')
```

```python
df_user.shape
```

```python
df_user = pd.read_csv('dataset/user.csv')
df_user = df_user.drop('Unnamed: 0',axis=1)
```
